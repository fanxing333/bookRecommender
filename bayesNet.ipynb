```python
# 第二层聚类
clusters_num = len(clusters)
for i in range(clusters_num):
    clusters[i].cutAndCount()
    # 返回词频最高的 top 个
    clusters[i].getTop(dimension=50)
    # 对每一个书名 title 进行向量化
    clusters[i].vectorize_titles()
    # 聚类
    clusters += clusters[i].cluster(classCount=second)
```

```python
names = [x.name for x in clusters]
# 制作数据表喂给贝叶斯估计器
rec_b = rec[["school", "title", "score"]].copy()
# 增加给类的列
for name in names:
    rec_b[name] = 0

# 如果分数大于0.3，则标为喜欢
for index, row in rec_b.iterrows():
    if row['score'] > 0.3:
        for i in range(first):
            exit_tag = False
            if row["title"] in clusters[i].titles.values:
                rec_b.loc[index, clusters[i].name] = 1
                
            for j in range(second):
                if row["title"] in clusters[first+i*second+j].titles.values:
                    rec_b.loc[index, clusters[first+i*second+j].name] = 1
                    exit_tag = True
                    break
                    
            if exit_tag:
                break
```

```python
from pgmpy.models import BayesianModel
from pgmpy.factors.discrete import TabularCPD
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
```

```python
nodes = []
for i, val in enumerate(names[:first]):
    relation = ('school', val)
    nodes.append(relation)
    for j in range(second):
        relation = (val, names[first+i*second+j])
        nodes.append(relation)
        
model = BayesianModel(nodes)
```

```python
from fuzzywuzzy import fuzz
def search(string, top=10):
    records = rec.copy()
    records['similarity'] = 0
    records = records.drop_duplicates(subset='title')[['title', 'author', 'similarity']].values

    for record in records:
        record[2] = fuzz.token_sort_ratio(string, record[0])

    return records[np.argsort(records[:, -1])][-top:][::-1]
```

```python
myschool = "信电学院"
keyword = "Python"
arr = search(keyword)
x = {}
for title in arr:
    for cluster in clusters[:first]:
        if title[0] in cluster.titles.values:
            if cluster.name not in x.keys():
                x[cluster.name] = 1
            else:
                x[cluster.name] += 1
            
            break

#print(arr)
#print(x)

cluster_class = sorted(x.items(), reverse = True)[0][0]
p = []
for cc in names[first:]:
    p.append(infer.query(variables=[cc], evidence={cluster_class: 1, "school": myschool}).values[1])
final_cluster = names[first + p.index(max(p))]

from Evaluation import Evaluation
eva = Evaluation(rec)
final_list = []
for i in clusters[names.index(final_cluster)].titles.values:
    e = eva.bookAvgEvaluate(i[0], myschool)
    score = e[0] * e[1]
    final_list.append([i[0], score])
    
print("---------模糊匹配----------")
print(arr)
print("---------个性推荐----------")
sorted(final_list, key=lambda x:x[1], reverse=True)[:10]
```
